```python
import os
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
nomask_path = "D:\\Code_uc\\face_reco\\output\\dataset_nomask"
mask_path = "D:\\Code_uc\\face_reco\\output\\dataset_mask"
augment_path = "D:\\Code_uc\\face_reco\\output\\augment_mask"
```

```python
gamma = 0.4
look_up_table = np.arange(256, dtype=np.float32)
look_up_table = (look_up_table/255.0)**gamma * 255
look_up_table = np.clip(look_up_table, 0, 255)
for subject_name in os.listdir(mask_path):
    subject_images_dir = os.path.join(mask_path, subject_name)
    new_subject_dir = os.path.join(augment_path, subject_name)
    os.makedirs(new_subject_dir)
    for i, img_name in enumerate(os.listdir(subject_images_dir)):
        img_path = os.path.join(subject_images_dir, img_name)
        img = cv.imread(img_path, cv.IMREAD_GRAYSCALE)
        img_1 = np.roll(img, 5, axis=1)
        img_2 = np.roll(img, -5, axis=1)
        img_3 = cv.LUT(img, look_up_table)
        img_3 = img_3.astype(np.uint8)
        cv.imwrite(os.path.join(new_subject_dir, f"{i}.jpg"), img)
        cv.imwrite(os.path.join(new_subject_dir, f"{i}_01.jpg"), img_1)
        cv.imwrite(os.path.join(new_subject_dir, f"{i}_02.jpg"), img_2)
        cv.imwrite(os.path.join(new_subject_dir, f"{i}_03.jpg"), img_3)
```
